```python
import requests
link_url = ' https://cdn.nba.com/static/json/liveData/scoreboard/todaysScoreboard_00.json'
headers  = {
    'Connection': 'keep-alive',
    'Accept': 'application/json, text/plain, */*',
    'x-nba-stats-token': 'true',
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36',
    'x-nba-stats-origin': 'stats',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-Mode': 'cors',
    'Referer': 'https://stats.nba.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
}
```

```python
r = requests.get(url=link_url, headers=headers).json()
r
```

```python
games = r['scoreboard']['games']
home_team = r['scoreboard']['games']
home_team
```

```python
df_todays_games = pd.DataFrame(games)
```

```python
for i, row in  df_todays_games.iterrows():
    df_todays_games['homeTeam'][i] = r['scoreboard']['games'][i]['homeTeam']['teamId']

for i, row in  df_todays_games.iterrows():
    df_todays_games['awayTeam'][i] = r['scoreboard']['games'][i]['awayTeam']['teamId']
df_todays_games
```

```python
df_todays_games = df_todays_games[['gameId', 'homeTeam', 'awayTeam','gameTimeUTC']]
```

```python
df_todays_games['gameTimeUTC'] = pd.to_datetime(df_todays_games['gameTimeUTC']).dt.strftime('%Y-%m-%d')
```

```python
df_todays_games
```

```python
df_todays_games.rename(columns={
    'gameId':'GAME_ID',
    'homeTeam': 'HOME_TEAM_ID',
    'awayTeam': 'AWAY_TEAM_ID',
    'gameTimeUTC' : 'GAME_DATE'
}, inplace=True)

df_todays_games
```

```python
df_todays_games['SEASON_YEAR'] = '2022-23'
df_todays_games= df_todays_games.set_index(['GAME_ID', 'SEASON_YEAR', 'GAME_DATE'])
```

```python
df_todays_games
```

```python
df_final_2 = pd.concat([df_final, df_todays_games], axis=0)
```

```python
df_final_2.index = df_final_2.index.set_levels([df_final_2.index.levels[0], df_final_2.index.levels[1], pd.to_datetime(df_final_2.index.levels[2])])
df_final_2=df_final_2.sort_index(level=2)
```

```python
df_final_2['HOME_WL'] = df_final_2['HOME_WL'].fillna(2)
df_final_2 = df_final_2.fillna(0)
df_final_2
```

```python
# Changing all columns to rolling averages using GroupBy and rolling
df_final_2 = df_final_2.groupby(by ='HOME_TEAM_ID', as_index= True).rolling(window = 20, on = 'HOME_WL').mean()
df_final_2.head(20)
```
