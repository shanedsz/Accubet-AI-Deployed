# This notebook will take all advanced data stats and prepare them for a Logistic Regression model 

```python
import pandas as pd
import numpy as np
from pathlib import Path
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
df_advanced_data = pd.read_csv(Path('original_data_advanced 2.csv'))
df_advanced_data.drop(columns='Unnamed: 0', inplace=True)
# Fixing Data types
df_advanced_data['GAME_DATE'] = pd.to_datetime(df_advanced_data['GAME_DATE'])
display(df_advanced_data.where(df_advanced_data['GAME_DATE']=='2023-03-24').value_counts().value_counts())
```

```python
display(df_advanced_data)
```

```python
df_home_teams = df_advanced_data.where(df_advanced_data['MATCHUP'].str.contains('vs')).dropna()
df_home_teams = df_home_teams.set_index(['GAME_ID','SEASON_YEAR', 'GAME_DATE'])
df_home_teams
```

```python
df_home_teams.drop(columns=['TEAM_ABBREVIATION','TEAM_NAME', 'MATCHUP'], inplace= True)
```

```python
df_home_teams.dtypes
df_home_teams['WL'].replace({'W': [1],
                             'L': [0]
},inplace=True)
```

```python
df_home_teams.columns = 'HOME_' + df_home_teams.columns
```

```python
df_home_teams['HOME_TEAM_ID'] = df_home_teams['HOME_TEAM_ID'].astype('int64')
```

```python
df_home_teams
```

```python
df_away_teams = df_advanced_data.where(df_advanced_data['MATCHUP'].str.contains('@')).dropna()
df_away_teams = df_away_teams.set_index(['GAME_ID', 'SEASON_YEAR', 'GAME_DATE'])
df_away_teams.drop(columns=['TEAM_ABBREVIATION','TEAM_NAME','MATCHUP', 'WL'], inplace=True)
```

```python
df_away_teams['TEAM_ID'] = df_away_teams['TEAM_ID'].astype('int64')
df_away_teams.columns = 'AWAY_' + df_away_teams.columns
```

```python
df_scoring_data = pd.read_csv(Path('original_data_scoring 3.csv'))
df_scoring_data.drop(columns=['Unnamed: 0', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'WL', 'MIN'], inplace=True)
df_scoring_data['GAME_DATE'] = pd.to_datetime(df_scoring_data['GAME_DATE'])
df_scoring_data = df_scoring_data.set_index(['GAME_ID', 'SEASON_YEAR', 'GAME_DATE'])
df_scoring_data
```

```python
df_home_teams_2 = df_scoring_data.where(df_scoring_data['MATCHUP'].str.contains('vs')).dropna()
```

```python
df_home_teams_2.drop(columns=['MATCHUP'],inplace= True)
```

```python
df_home_teams_2.columns = 'HOME_' + df_home_teams_2.columns
```

```python
df_home_teams_2.columns
```

```python
df_home_teams_2['HOME_TEAM_ID']= df_home_teams_2['HOME_TEAM_ID'].astype('int64')
```

```python
df_away_teams_2 = df_scoring_data.where(df_scoring_data['MATCHUP'].str.contains('@')).dropna()
df_away_teams_2.drop(columns=['MATCHUP'], inplace=True)
df_away_teams_2['TEAM_ID'] = df_away_teams_2['TEAM_ID'].astype('int64')
```

```python
df_away_teams_2.columns = 'AWAY_' + df_away_teams_2.columns
df_away_teams_2
```

```python
df_concat_2 = pd.concat([df_home_teams_2, df_away_teams_2], axis=1)
df_concat_2.columns
```

```python
df_four_factors = pd.read_csv(Path('original_data_four_factors 3.csv'))
df_four_factors.drop(columns=['Unnamed: 0', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'WL', 'MIN'], inplace=True)
df_four_factors['GAME_DATE'] = pd.to_datetime(df_four_factors['GAME_DATE'])
df_four_factors = df_four_factors.set_index(['GAME_ID', 'SEASON_YEAR', 'GAME_DATE'])
df_four_factors
```

```python
df_home_teams_3 = df_four_factors.where(df_four_factors['MATCHUP'].str.contains('vs')).dropna()
df_home_teams_3.drop(columns=['MATCHUP'],inplace= True)
df_home_teams_3.columns = 'HOME_' + df_home_teams_3.columns
df_home_teams_3['HOME_TEAM_ID']= df_home_teams_3['HOME_TEAM_ID'].astype('int64')
```

```python
df_away_teams_3 = df_four_factors.where(df_four_factors['MATCHUP'].str.contains('@')).dropna()
df_away_teams_3.drop(columns=['MATCHUP'], inplace=True)
df_away_teams_3['TEAM_ID'] = df_away_teams_3['TEAM_ID'].astype('int64')
df_away_teams_3.columns = 'AWAY_' + df_away_teams_3.columns
```

```python
df_concat_3 = pd.concat([df_home_teams_3, df_away_teams_3], axis=1)
df_concat_3
```

```python
df_final = pd.concat([df_concat, df_concat_2, df_concat_3], axis =1 )
```

```python
df_final = df_final.loc[:,~df_final.columns.duplicated()].copy()
```

```python
df_wl = df_final['HOME_WL']
df_wl
```
